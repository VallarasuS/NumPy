## Operations

Arithmetic operations on NumPy arrays are applied element wise, and returns a 'NEW' array with results.


```python
import numpy as np

a = np.array([10,20,30,40])
b = a + 1
print(a)
print(b)

c = b - a
print(c)
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([1, 2, 3, 4])

z = x @ y
print(z)
```

Some short hand operations like `*=` , `+=` modify the source array in place rather than creating a new one


```python
a_ones = np.ones((2,3))
print(a_ones)

a_ones += 1
print(a_ones)
```

```python
scores = np.array([50,45,30,49,62])
results = scores > 35

print("Pass / Faile ? :", results)
print("Centum", scores == 0)
```

## Math Operations

- sum, prod
- min, max
- cumsum, cumprod
- mean / average, median

for more math operations check [here](https://numpy.org/devdocs/user/quickstart.html#basic-operations)

```python
scores = np.array([1, 2, 3, 4, 5])
total = scores.sum()
print(total)

low = scores.min()
print(low)

high = scores.max()
print(high)

fact = scores.prod()
print(fact)

cum_sum = scores.cumsum()
print(cum_sum)

cum_prod = scores.cumprod()
print(cum_prod)
```
