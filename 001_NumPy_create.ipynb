## NumPy Intro

It is a table of elements (usually numbers), all of the same type, indexed by a tuple of non-negative integers. In NumPy dimensions are called axes.

```python
[[1., 0., 0.],
 [0., 1., 2.]]
```

In NumPy dimensions are called axes. In this example the array has 2 axes. The first axis (column) has a length of 2, second axes has a length of 3.

### Attributes

NumPy's array is called ndarray, most important attributes are

### ndarray.ndim

number of axes (dimesions) of array

### ndarray.shape
dimentions of the array. tuple of int indicating the size of array in each dimension. matrix with n rows and m columns has a `shape` of `(n,m)`.
The length of `shape` is therefore the number of axis `ndim`

### ndarray.size

total number of elements of the array, equal to the product of the elements of `shape`

### ndarray.dtype
type of elements in the array

### ndarray.itemsize
size in bytes of each element of the array


### ndarray.data
buffer containing elements

## Array Creation - np.array

can creaet an array from python list or tuple. type of array will be type of elements in the sequence.

```python
import numpy as np

a = np.array([1, 2, 3, 4])
print(a)
print(type(a))
```

```python
print("ndim", a.ndim)
print("shape", a.shape)
print("size", a.size)
```

```python
a2d = np.array([(1,2,3), (4,5,6)])
print(a2d)

print("ndim", a2d.ndim)
print("shape", a2d.shape)
print("size", a2d.size)
```

## Array creation - np.zeroes

often array elements are unknow at creation, numpy offers several ways to create arrays with placeholders

- zeros
- ones
- empty
- arange

other ways to [create](https://numpy.org/devdocs/reference/routines.array-creation.html) ndarray

```python
aZeros = np.zeros((2,5))
print(aZeros)
```

```python
aOnes = np.ones((2,2,4))
print(aOnes)
```

```python
a_range = np.arange(6)
print(a_range)
```

```python
a_range = np.arange(2,10,2)
print(a_range)
```
