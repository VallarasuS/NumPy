## Array Shape Manipulation

all return a modified array, but do not change the original array

- ravel
- reshape
- resize (mutate)


```python
# Ravel

import numpy as np

a = np.array([[1,2,3,4], [5,6,7,8]])
print(a)

a1D = a.ravel()
print(a1D)
```

```python
# Re-shape

a1D = np.arange(12)
print(a1D)

a2D = a1D.reshape(4,3)
print(a2D)

a3D = a1D.reshape((2,2,3))
print(a3D)
```

```python
# Resize

a1D = np.arange(20)
print(a1D)


a1D.resize((2,2,5))
print(a1D)
```

## Stacking

- hstack
- vstack

```python
# horizontal stack

x = np.array([1, 2, 3, 4])
y = np.array([5, 6, 7, 8])

z = np.vstack((x, y))
print(z)
```

```python
# vertical stack

x = np.arange(10).reshape(2,5)
print(x)

y = np.ones((2,2), dtype="int32")
print(y)

z = np.hstack((x, y))
print(z)
```

## Split Arrays

- hsplit
- vslpit

```python
a = np.arange(20)
print(a)

xs = np.hsplit(a, 4)
[a, b, c, d]  = xs
print(a)
print(d)
```

```python
a = np.arange(20).reshape(4,5)
print(a)

[x, y] = np.vsplit(a, 2)
print(x)
print(y)
```
